```python
import requests

url = "https://adventofcode.com/2020/day/2/input"
response = requests.get(url)
print(f"The secret message is: {response.text}")
```

```python
path = 'C:\\Users\Agne\Programavimas\AdventOfCode\inputDay2.txt'
input_file = open(path,'r')

correct = 0

for line in input_file:
    print(line)
    minimal = int(line.split('-')[0])
    maximum = int(line.split('-')[1].split(' ')[0])
    letter = line.split(' ')[1].replace(':', '')
    password = line.split(' ')[2]
    count = password.count(letter)
    if (count >= minimal and count <= maximum):
        correct += 1
        
print(correct)
    
```

```python
 # part 2
    
path = 'C:\\Users\Agne\Programavimas\AdventOfCode\inputDay2.txt'
input_file = open(path,'r')

correct = 0

for line in input_file:
    print(line)
    index1 = int(line.split('-')[0]) - 1
    index2 = int(line.split('-')[1].split(' ')[0]) -1
    letter = line.split(' ')[1].replace(':', '')
    password = line.split(' ')[2]
    if (password[index1] == letter and password[index2] != letter):
        correct += 1
    elif (password[index2] == letter and password[index1] != letter):
        correct += 1
        
print(correct)
```
