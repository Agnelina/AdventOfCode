## Day 4

```python
path = 'C:\\Users\Agne\Programavimas\AdventOfCode\inputDay4.txt'
input_file = open(path,'r')

keys = ["byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"] # "cid"

validCount = 0

currentPassKeys = []

def isValid(requiredKeys, givenKeys):
    for key in requiredKeys:
        if key not in givenKeys:
            return False
    return True

for line in input_file:
    if (line == "\n"):
        if (isValid(keys, currentPassKeys)):
            validCount += 1
        currentPassKeys.clear()
    else:       
        items = line.split(" ")
        for item in items:
            currentPassKeys.append(item.split(":")[0])

if (isValid(keys, currentPassKeys)):
    validCount += 1
currentPassKeys.clear()
            
print(validCount)
input_file.close()
```

```python
import re

# other regex example:
#  text = input("enter:")
#  regex = r"([0-9a-zA-Z]+)"
#  match = re.match(regex, text)

path = 'C:\\Users\Agne\Programavimas\AdventOfCode\inputDay4.txt'
input_file = open(path,'r')

validCount = 0

currentPass = {
    "byr":"",
    "iyr":"",
    "eyr":"",
    "hgt":"",
    "hcl":"",
    "ecl":"",
    "pid":""
}

def isValid(passport):
    for key in passport:
        if (passport[key] == ""):
            return False
        
    birthYear = int(passport["byr"])
    issueYear = int(passport["iyr"])
    expirationYear = int(passport["eyr"])
    if (len(passport["hgt"]) < 4 ):
        return False
    heightUnit = passport["hgt"][-2:]
    heightValue = int(passport["hgt"][0:-2]) # jie ilgis maiau nei 2 problema
    hairColor = passport["hcl"]
    hairPattern = r"#{1}^[0-9a-f]{6}"
    eyeColor = passport["ecl"]
    eyeOptions = ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"]
    passportID = passport["pid"]
    pidPattern = r"[0-9a-f]{9}"
    
    if ((birthYear >= 1920 and birthYear <= 2002)
        and (issueYear >= 2010 and issueYear <= 2020)
        and (expirationYear >= 2020 and expirationYear <= 2030)
        and ((heightUnit == "cm" and heightValue >= 150 and heightValue <= 193) 
          or(heightUnit =="in" and heightValue >= 59 and heightValue <= 76))
        and (re.fullmatch(r"^#{1}[0-9a-f]{6}", hairColor) is not None)
        and (eyeColor in eyeOptions )
        and (re.fullmatch(r"[0-9a-f]{9}", passportID) is not None)
       ):
        return True
    else:
        return False

for line in input_file:
    if (line == "\n"):
        if (isValid(currentPass)):
            validCount += 1
        currentPass = {x: "" for x in currentPass}
    else:       
        items = line.split(" ")
        for item in items:
            if (item.split(":")[0] != "cid"):
                currentPass[item.split(":")[0]]= item.split(":")[1].replace("\n", '')

if (isValid(currentPass)):
    validCount += 1
            
print(validCount)
input_file.close()
```

## Day 5

```python
path = 'C:\\Users\Agne\Programavimas\AdventOfCode\inputDay5.txt'
input_file = open(path,'r')

maxSeatID = 0
seats = []

for line in input_file:
    row = []
    for n in range(128):
        row.append(n)
    for x in range(7):
        if line[x] == "B":
            row = row[len(row)//2:]
        else:
            row = row[:len(row)//2]
        
    column = []
    for n in range(8):
        column.append(n)
    for x in range(3):
        if line[7+x] == "R":
            column = column[len(column)//2:]
        else:
            column = column[:len(column)//2]
    
    seatID = row[0]*8+column[0]
    seats.append(seatID)
    if seatID > maxSeatID:
        maxSeatID = seatID
        
seats.sort()
for i in range(len(seats)-1):
    if seats[i]+1 != seats[i+1]:
        print(seats[i])
        print(seats[i+1])
        break

print(maxSeatID)

input_file.close()
```

## Day 6

```python
path = 'C:\\Users\Agne\Programavimas\AdventOfCode\\2020\inputDay6.txt'
input_file = open(path,'r')

# part 1: anyone
count = 0

groupAnswerSet = set()

for line in input_file:
    if (line == "\n"):
        count += len(groupAnswerSet)
        groupAnswerSet.clear()
    else:
        for char in line:
            if char != "\n":
                groupAnswerSet.add(char)
    
count += len(groupAnswerSet)
print(count)
    
input_file.close()
```

```python
# part 2: everyone

from string import ascii_lowercase

path = 'C:\\Users\Agne\Programavimas\AdventOfCode\\2020\inputDay6.txt'
input_file = open(path,'r')

countEveryone = 0

groupAnswers = {}

for c in ascii_lowercase:
    groupAnswers[c] = 0

groupSize = 0

for line in input_file:
    if (line == "\n"):
        for c in ascii_lowercase:
            if groupAnswers[c] == groupSize:
                countEveryone += 1
        groupSize = 0
        for c in ascii_lowercase:
            groupAnswers[c] = 0
    else:
        groupSize +=1
        for char in line:
            if char != "\n":
                groupAnswers[char] += 1
                
for c in ascii_lowercase:
    if groupAnswers[c] == groupSize:
        countEveryone += 1

print(countEveryone)
    
input_file.close()
```
